# Data Exploration

Imports

```python
import pandas
import matplotlib.pyplot as plotter
```

Import data preprocessing libraries

```python
from sklearn.preprocessing import MinMaxScaler
```

Import dataframes from csv

```python
cpu_df = pandas.read_csv("no_app_cpu.csv")
network_tx_df = pandas.read_csv("no_app_network_tx.csv")
network_rx_df = pandas.read_csv("no_app_network_rx.csv")
```

```python
cpu_df.describe()
```

```python
network_rx_df.describe()
```

```python
network_tx_df.describe()
```

```python
df = pandas.concat([cpu_df, network_tx_df, network_rx_df], join='inner', axis=1)
```

```python
df.describe()
```

```python
df.shape
```

Remove index and time columns

```python
df = df.drop(['Unnamed: 0','Time'],1).fillna(value=0)
```

### Scale data

```python
scaler = MinMaxScaler()
scaler.fit(df.as_matrix())
```

```python
data = scaler.transform(df).T
```

```python
data.shape
```

```python
for series in range(data.shape[0]):
    plotter.plot(data[series,1000:1100])
plotter.show()
```

```python
for series in range(data.shape[0]):
    plotter.plot(data[series,10000:10100])
plotter.show()
```

```python
plotter.imshow(data[:,1000:1500])
plotter.show()
```
