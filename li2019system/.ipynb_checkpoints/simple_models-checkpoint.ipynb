### Imports

```python
import numpy

from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import classification_report
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plotter
```

## Import data into arrays

```python
bcast_data = numpy.load('bcast_arr.npy')
noapp_data = numpy.load('noapp_arr.npy')
```

```python
bcast_data.shape
```

```python
noapp_data.shape
```

### Create labels

```python
bcast_y = numpy.zeros(bcast_data.shape[1])
noapp_y = numpy.zeros(noapp_data.shape[1]) + 1.0
```

```python
bcast_y.shape
bcast_y
```

```python
noapp_y.shape
noapp_y
```

```python
labels = numpy.hstack((bcast_y,noapp_y))
```

```python
labels.shape
```

```python
labels
```

```python
data = numpy.hstack((bcast_data,noapp_data)).T
```

```python
data.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.3, random_state=4)
```

```python
y_train
```

```python
y_test
```

```python
lr = SGDClassifier(max_iter=1000, verbose=2, loss='perceptron')
```

```python
lr.fit(X_train, y_train)
```

```python
lr.score(X_test, y_test)
```

```python
pred = lr.predict(X_test)
print(classification_report(y_test, pred, target_names=['bcast','noapp']))
```

```python
pca = PCA(n_components=2)
c_data = pca.fit_transform(data)
plotter.scatter(c_data[:,0], c_data[:,1])
```
