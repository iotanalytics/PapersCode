## Seperate into proper dataframes for each node

```python
dframes = [cpu_df, txnet_df, rxnet_df, context_df, disk_df]
node = {}

for i in range(1,5):
    frames = []
    
    for dframe in dframes:
        columns = list(filter(lambda x: f'bb{i}l' in x, dframe.columns))
        frames.append(dframe[columns])

    node[i] = pandas.concat(frames, join='inner', axis=1).fillna(0)[:38200]
```

```python
for i in range(1,5):
    print(node[i].shape)
```

## Get data

```python
data_matrices = []

for i in range(1,5):
    data_matrices.append(node[i].as_matrix())
    
data = numpy.vstack(data_matrices)
```

```python
data.shape
```

## Prepare scaler

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler
scaler = MinMaxScaler()
```

```python
scaler.fit(data)
del data
```

---

## Correrlation measurement

```python
bb1 = scaler.transform(data_matrices[0])
bb2 = scaler.transform(data_matrices[1])
bb3 = scaler.transform(data_matrices[2])
bb4 = scaler.transform(data_matrices[3])
```

```python
DIM = bb1.shape[1]
print(DIM)
```

```python
series = [('12', bb1, bb2), ('13', bb1, bb3), ('14', bb1, bb4), ('23', bb2, bb3), 
          ('24', bb2, bb4), ('34', bb3, bb4), ('TD12', numpy.vstack((bb1[33:48,:], bb1[5:,:])), bb2)]

for m in series:
    print(f"For {m[0]}")
    corr = numpy.zeros((DIM,DIM))
    for i in range(DIM):
        for j in range(DIM):
            corr[i][j] = pearsonr(m[1][i],m[2][j])[0]
    #print(corr)
    plotter.imshow(corr)
    plotter.show()
```

## Reshape matrices for training

```python
TIME_STEPS = 20

for i in range(len(data_matrices)):
    
    transformed = scaler.transform(data_matrices[i])
    
    stack = []
    for j in range(transformed.shape[0]-TIME_STEPS):
        
        stack.append(transformed[j:j+TIME_STEPS,:])
        
    stack = numpy.transpose(numpy.stack(stack, axis=2), (2,0,1))
    print(stack.shape)
    DIM = stack.shape[2]
    #transformed = transformed.reshape((-1, TIME_STEPS, DIM))
    data_matrices[i] = stack
```

## Prepare FC Autoencoder

### Flatten dataset

```python
labels = []

for i in range(len(data_matrices)):
    
    matrix = data_matrices[i]
    BATCH = matrix.shape[0]
    
    label = numpy.zeros(BATCH)
    
    if i == 3:
        label += 1.0
        
    labels.append(label)
    
    data_matrices[i] = matrix.reshape((BATCH, -1))
    
    print(f"Matrix {i} has MAX: {numpy.max(data_matrices[i])} MIN: {numpy.min(data_matrices[i])} and shape {data_matrices[i].shape}")
```

```python
X = numpy.vstack(data_matrices)
y = numpy.hstack(labels)
print(y)
```

### Prepare train test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.3)
```

### Prepare model

```python
from keras.models import Model
from keras.layers import Dense, Input
from keras.callbacks import EarlyStopping
```

### Simple AE

```python
input_layer = Input(shape=(TIME_STEPS * DIM,))
encoder = Dense(100, activation='relu')(input_layer)
#encoder = Dense(50, activation='relu')(encoder)
encoder = Dense(4, activation='relu')(encoder)
```

```python
#decoder = Dense(50, activation='relu')(encoder)
decoder = Dense(100, activation='relu')(encoder)
output = Dense(TIME_STEPS * DIM, activation='sigmoid')(decoder)
```

### train model

```python
encoder_model = Model(input_layer, encoder)
model = Model(input_layer, output)
```

```python
model.compile(loss='mean_squared_error', metrics=['mse'], optimizer='adam')
```

```python
estopper = EarlyStopping(min_delta=0.00001, patience=5, monitor='val_loss')
model.fit(X_train, X_train, epochs=10000, validation_split=0.1, callbacks=[estopper], shuffle=False)
```

```python
X_train = encoder_model.predict(X_train)
X_test = encoder_model.predict(X_test)
```

### Cluster based on reduced dimensionality

```python
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
```

```python
for i in range(1):
    print(f"-----Running experiment {i}-----")
    for i in range(1,5):
        kmeans = KMeans(n_clusters=i)
        kmeans.fit(X_train, y_train)
        print(f"for n = {i} accuracy is {accuracy_score(y_test, kmeans.predict(X_test))}")
        plotter.scatter(X_test[:,2], X_test[:,1], c=kmeans.predict(X_test), alpha=0.3)
        plotter.show()
```

## Recurrent AE

```python
X = numpy.vstack(data_matrices[:-1])
X.shape
```

### Train test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test = train_test_split(X, test_size=0.1, shuffle=False)
```

### Build recurrent AE

```python
from keras.models import Model
from keras.layers import Input, GRU, Dense, RepeatVector
from keras.callbacks import EarlyStopping
```

```python
input_layer = Input(shape=(TIME_STEPS, DIM))
#encoding = GRU(10, activation='relu', return_sequences=True)(input_layer)
encoding = GRU(3, activation='relu')(input_layer)
```

```python
decoded = RepeatVector(TIME_STEPS)(encoding)
#decoded = GRU(10, activation='relu', return_sequences=True)(decoded)
output = GRU(DIM, activation='sigmoid', return_sequences=True)(decoded)
```

```python
model = Model(input_layer, output)
encoder = Model(input_layer, encoding)
```

```python
estopper = EarlyStopping(min_delta=0.00001, patience=5, monitor='val_loss')
model.compile(loss='mean_squared_error', metrics=['mse'], optimizer='rmsprop')
model.fit(X_train, X_train, epochs=10000, validation_split=0.1, callbacks=[estopper], shuffle=False)
```

```python
bb1 = encoder.predict(data_matrices[0])
bb2 = encoder.predict(data_matrices[1])
bb3 = encoder.predict(data_matrices[2])
bb4 = encoder.predict(data_matrices[3])
```

```python
plotter.scatter(bb1[:,0], bb1[:,1], c='r', marker='.', alpha=0.1)
plotter.scatter(bb2[:,0], bb2[:,1], c='r', marker='.', alpha=0.1)
plotter.scatter(bb3[:,0], bb3[:,1], c='r', marker='.', alpha=0.1)
plotter.scatter(bb4[:,0], bb4[:,1], c='y', marker='.', alpha=0.1)
plotter.show()
```

```python
from sklearn.mixture import GaussianMixture
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import classification_report
```

```python
X = numpy.vstack([bb1, bb2, bb3, bb4])
y = numpy.hstack([numpy.zeros(bb1.shape[0]), numpy.zeros(bb2.shape[0]), numpy.zeros(bb3.shape[0]), 1.0+numpy.zeros(bb4.shape[0])])

X_train, X_test, y_train, y_test = train_test_split(X,y)
#X_train = encoder.predict(X_train)
#X_test = encoder.predict(X_test)
```

```python
gmm = GaussianMixture(n_components=2)
gmm.fit(X_train)

plotter.scatter(X_test[:,1], X_test[:,0], c=gmm.predict(X_test), alpha=0.3)
plotter.show()

agg = AgglomerativeClustering(n_clusters=2)
agg.fit(X_train)

plotter.scatter(X_test[:,1], X_test[:,0], c=agg.predict(X_test), alpha=0.3)
plotter.show()

print(classification_report(y_pred=gmm.predict(X_test), y_true=y_test))
print(classification_report(y_pred=agg.predict(X_test), y_true=y_test))
```
