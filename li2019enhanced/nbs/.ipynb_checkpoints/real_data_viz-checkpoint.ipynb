Try filter design

```python
order = 2
cutoff = 0.1

B,A = butter(order, cutoff, output="ba")
```

```python
filtered = filtfilt(B, A, numpy.copy(power))
```

```python
plotter.plot(power[:100], alpha=0.9, linewidth=0.5, label="power")
plotter.plot(filtered[:100], alpha=0.9, linewidth=0.5, label="filtered")
plotter.legend()
plotter.show()

res = power - filtered
plotter.plot(res[:100], alpha=0.9, linewidth=0.5, label="residual")
plotter.show()
```
