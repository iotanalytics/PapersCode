## Seperate into proper dataframes for each node

```python
dframes = [cpu_df, txnet_df, rxnet_df, context_df, disk_df]
node = {}

for i in range(1,5):
    frames = []
    
    for dframe in dframes:
        columns = list(filter(lambda x: f'bb{i}l' in x, dframe.columns))
        frames.append(dframe[columns])

    node[i] = pandas.concat(frames, join='inner', axis=1).fillna(0)[:38200]
```

```python
for i in range(1,5):
    print(node[i].shape)
    
print(node[1].columns)
```

## Get data

```python
data_matrices = []

for i in range(1,5):
    data_matrices.append(node[i].as_matrix())
    
data = numpy.vstack(data_matrices)
```

```python
X = numpy.transpose(X, (1, 0, 2))
train_X = train_X
train_X = numpy.transpose(train_X, (1, 0, 2))
#X = X.reshape((-1,382,29))
val_X = numpy.transpose(val_X, (1, 0, 2))
test_X = numpy.transpose(test_X, (1, 0, 2))
#val_X = val_X.reshape((-1,382,29))
print(X.shape)
print(val_X.shape)
```

```python
def flat_generator(X, tsteps = 5):
    i = 0
    
    while True:
        batch_X = X[:,i:i+tsteps,:]
        batch_y = X[:,i+tsteps,:]
            
        #batch_X = batch_X.reshape((batch_X.shape[0], -1))
        #print(batch_X.shape)
        #print(batch_y.shape)
        
        yield batch_X, batch_y
        
        i += 1
        if i > (X.shape[1] - tsteps - 1):
            i = 0
            continue
```

```python
TIMESTEPS = 20
DIM = 29
tgen = flat_generator(X, TIMESTEPS)
vgen = flat_generator(val_X, TIMESTEPS)
```

```python
cswitch = X[0][:,0]
plotter.plot(cswitch[35500:35700], alpha=0.5)
plotter.show()
```

```python
f0, p = periodogram(cswitch[35480:35500])
plotter.semilogy(f0, p, 'g-', alpha=0.3)


f1, p = periodogram(cswitch[35500:35520])
plotter.semilogy(f1, p, 'g-', alpha=0.3)


f2, p = periodogram(cswitch[35520:35540])
plotter.semilogy(f2, p, 'y-', alpha=0.3)


f3, p = periodogram(cswitch[35540:35560])
plotter.semilogy(f3, p, 'r-', alpha=0.3)


f4, p = periodogram(cswitch[35560:35580])
plotter.semilogy(f4, p, 'r-', alpha=0.3)
plotter.show()
```

```python
for i in range(0,len(cswitch)-20,20):
    #print((i,i+20))
    f, p = periodogram(cswitch[i:i+20])
    
    if i >= 35520 and i <= 35600:
        plotter.semilogy(f,p,'r-',alpha=0.1)
    else:
        plotter.semilogy(f,p,'g-',alpha=0.1)

plotter.show()
```
